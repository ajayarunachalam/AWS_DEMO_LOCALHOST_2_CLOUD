```python
# Import libraries and packages
from sklearn.datasets import load_boston
import numpy as np
```

```python
# Load Boston House Price Data
X, y = load_boston(return_X_y=True)
```

```python
print(f' Predictors: {X}')
```

```python
print('Outcome: {}' .format(y))
```

```python
print(X.shape)
print(X)

print(y.shape)
print(y)
```

```python
import xgboost as xgb
print(xgb.__version__)
```

```python
# Split Train, Test data sets for modeling

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
print('Data shape:')
print('X_train: %s, X_test: %s \n' %(X_train.shape, X_test.shape))
```

```python
# Train a XGBoost Regression Model
bt = xgb.XGBRegressor(
    max_depth=5, learning_rate=0.2, n_estimators=10, objective="reg:squarederror"
)  # Setup xgboost model
```

```python
bt.fit(X_train, y_train, verbose=False)  # Train it to our data
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
# Predict and compare with real labels
bt.predict(X_test)
```

```python
# Save the model as a file using joblib dump

model_file_name = "DEMO-local-xgboost-reg-model"
```

```python
# Use joblib which is considered better than pickling 
import joblib
joblib.dump(bt, model_file_name) 
```
